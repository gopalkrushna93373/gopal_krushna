```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\gopal\OneDrive\Desktop\Amazon E_Commerce Sales Analysis\cleaned_Amzon_sales.csv")
df.head(10)
```

1. Total Revenue

```python
total_revenue = df["Amount"].sum()
print(f"Total Revenue : Rs.{total_revenue:,.2f}")
```

2. Total Orders

```python
Total_orders = len(df)
print("Total order :", Total_orders)
```

3. Total Quantity sold :

```python
total_quantity = df["Qty"].sum()
print("Total Quantity Sold :-", total_quantity)
```

4. Average Order Value :

```python
avg_order_value = df["Amount"].mean()
print(f"Average_Order_Value :- Rs.{avg_order_value:.0f}")
```

5. Maximum Order Value :

```python
max_order = df["Amount"].max()
print(f"Maximum order value :- Rs.{max_order:.0f}")
```

6. Cancellation Rate

```python
lenght = len(df)
cancel_order = df["Status"].str.contains(
    "Cancelled",
    case=False,
    na=False
).sum()
cancel_rate = (cancel_order/lenght)*100
print(f"Cancellatio Rate : {cancel_rate:.0f}%")
```

7. Delivery rate

```python
delivered = df["Courier Status"].str.contains(
    "Shipped",
    case=False,
    na=False
    
).sum()
delivery_rate = (delivered/len(df))*100
print(f"Delivery Rate :- {delivery_rate:.2f}%")
```

8. Top 10 Categories by Revenue

```python
top_catrgories = df.groupby("Category")["Amount"].sum().nlargest(10).reset_index(name="Total Revenue")
top_catrgories
```

9. Top 10 States by revenue

```python
top_states = df.groupby("ship-state")["Amount"].sum().nlargest(10).reset_index(name="Total_Revenue")
top_states
```

10. Top 10 states and cities total revenue

```python
top_city_states_revenue = df.groupby(["ship-state", "ship-city"])["Amount"].sum().nlargest(10).reset_index(name="total_revenue")
top_city_states_revenue
```

11. Revenue by Size

```python
size_wise_revenue = df.groupby("Size")["Amount"].sum().reset_index(name="total_revenue")
size_wise_revenue
```

12. Revenue by Fulfilment

```python
fulfilment_sales = df.groupby("Fulfilment")["Amount"].sum().reset_index(name="total_revenue")
fulfilment_sales
```

13. Revenue by Courier Status

```python
courier_sales = df.groupby("Courier Status")["Amount"].sum().reset_index(name="Total Qty")
courier_sales
```

14. find states whose revenue is greater then avg revenue

```python
state_revenue = df.groupby("ship-state")["Amount"].sum().reset_index(name="Total Rvn")
avg_revenue = state_revenue["Total Rvn"].mean()
above_avg_rvn = state_revenue[state_revenue["Total Rvn"] > avg_revenue].sort_values(by="Total Rvn", ascending=False)
#print("Average Revenue :", round(avg_revenue, 2))
print(above_avg_rvn)
```

15. Total Shipped revenue

18. Monthly and day wise total revenue

19. Weekend total sales

```python
df[df["Weekend"]==True]["Amount"].sum()
```

```python
df.groupby("Weekend")["Amount"].sum()
```

20. state wise KPIs

```python
state_kpi = df.groupby("ship-state").agg({
    "Amount" : ["sum", "mean"],
    "Qty" : "sum"
})
state_kpi
```
